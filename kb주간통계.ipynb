```python
import pandas as pd
import plotly.express as px
```

```python
fileNmae = '20230320_주간시계열.xlsx'
df = pd.read_excel('./data/' + fileNmae, sheet_name='7.매수매도', index_col=0, header=[1, 2], skiprows=[3], skipfooter=6 )
df.head(3)
```

```python
data = df.stack(level=(0,1)).reset_index()
data.columns = ['기준시점', '지역', '구분', '매수매도']
data['지역'] = data['지역'].apply(lambda x : x.split(' ')[0])

region = [ '서울특별시', '경기도', '인천광역시', '부산광역시', '대구광역시', '대전광역시', '광주광역시', '울산광역시', '세종특별자치시']
data = data.loc[data['지역'].isin(region)]
data.head(10)
```

```python
buysell_order = ['매도자많음', '매수자많음', '매수우위지수']

fig = px.line(data, x='기준시점', y='매수매도', color='구분',  facet_col='지역', facet_col_wrap=3, \
            category_orders={'지역': region, '구분' : buysell_order},   \
            title='KB부동산 매수매도 지수', width=1200, height=1000)
fig.for_each_xaxis(lambda xaxis: xaxis.update(showticklabels=True))

fig.show()
```
